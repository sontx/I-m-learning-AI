### Kiểu dữ liệu
- age: continuous numerical -> rỏ ràng là số nguyên liên tiếp trong khoản 1-100?? và có thể tính các phép toán như min/max...
- job: nominal categorical -> text nên là category, các job ko liên quan tới nhau nên chắc chắn ko phải là oridinal rồi
- marital: nominal categorical -> text nên là category, tình trạng hôn nhân chỉ có 2 trạng thái như sex mà thôi
- education: oridinal categorical -> là text nên là category, các cấp độ giáo dục nó kế thừa và liên kết với nhau
- balance: continuous numerical -> là số và có thể tính toán các phép toán học lên đc
- housing: nominal categorical -> cái này lại giống marital rồi
- loan: giống bên trên
- contact: nominal categorical -> là text và các giá trị của nó ko liên quan j tới nhau
- day: continues numerical -> có thể tính đc ngày liên lạc sớm nhất, muộn nhất hay trung bình....
- month: continues numerical -> căng vl :| nó là text nhưng mà t có thể tính đc các giá trị kiểu như tháng liên lạc sớm nhất, muộn nhất hay trung bình liên lạc vào tháng nào, loạn vl

```python
from sklearn.impute import SimpleImputer
X=df.values
```

```python
job_impute= SimpleImputer(np.nan, strategy="most_frequent")
X[:,1]=job_impute.fit_transform(X[:,1].reshape(-1,1)).ravel()

marital_impute= SimpleImputer(np.nan, strategy="most_frequent")
X[:,2]=marital_impute.fit_transform(X[:,2].reshape(-1,1)).ravel()

education_impute= SimpleImputer(np.nan, strategy="most_frequent")
X[:,3]=education_impute.fit_transform(X[:,3].reshape(-1,1)).ravel()

housing_impute= SimpleImputer(np.nan, strategy="most_frequent")
X[:,5]=housing_impute.fit_transform(X[:,5].reshape(-1,1)).ravel()

contact_impute= SimpleImputer(np.nan, strategy="most_frequent")
X[:,7]=contact_impute.fit_transform(X[:,7].reshape(-1,1)).ravel()

month_impute= SimpleImputer(np.nan, strategy="most_frequent")
X[:,9]=month_impute.fit_transform(X[:,9].reshape(-1,1)).ravel()
```

```python
from sklearn.preprocessing import LabelEncoder
```
